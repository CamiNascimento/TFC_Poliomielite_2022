# Gerando base para criação do mapa de cobertura QGIS

```python
#Montando a base de dados para gerar o mapa de cobertura no QGIS
dfRE = pd.read_csv("MunRegioes.csv", index_col=1)
dfmapa=dfRE
dfmapa=dfmapa.set_index("Município")
#print(dfmapa)

dfRE=dfRE.sort_values(by=['Média Cob Vac'])
#print(dfRE)

q1 = dfRE["Média Cob Vac"].quantile(.33)
q2 = dfRE["Média Cob Vac"].quantile(.66)

dfRE= dfRE[["Média Cob Vac"]]
#print(dfRE)

print(q1)
print(q2)

LV=[]
LB=[]
LM=[]
LA=[]
for i in dfmapa["Média Cob Vac"]:
    #print(i)
    if i not in LV:
        LV.append(i)
        if i<q1:
            LB.append(dfmapa.index[dfmapa['Média Cob Vac']==i].tolist())
        elif i>=q1 and i<q2:
            LM.append(dfmapa.index[dfmapa['Média Cob Vac']==i].tolist())
        else:
            LA.append(dfmapa.index[dfmapa['Média Cob Vac']==i].tolist())

lct=[]
lmt=[]
for i in LB:
    for j in i:
        lmt.append(j)
        lct.append("baixa")
    
for i in LM:
    for j in i:
        lmt.append(j)
        lct.append("média")

for i in LA:
    for j in i:
        lmt.append(j)
        lct.append("alta")

 
dicionário = {'Município':lmt ,'TaxaCobertura': lct}
dfMapa = pd.DataFrame(dicionário)
dfMapa=dfMapa.set_index("Município")
print(dfMapa)

NomeExcel = pd.ExcelWriter('CobMapa.xlsx')
dfMapa.to_excel(NomeExcel)
NomeExcel.save()
```
